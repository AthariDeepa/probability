```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
mean = 50
std_dev = 54
num_samples = 1000
data = np.random.normal(mean, std_dev, num_samples)

plt.figure(figsize=(10, 5))
sns.histplot(data, bins=30, kde=True, color='blue', stat='density', label="Histogram")

x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)
pdf = norm.pdf(x, mean, std_dev)
plt.plot(x, pdf, 'r', label="PDF", linewidth=2)

plt.xlabel("Value")
plt.ylabel("Density")
plt.title("Histogram and Probability Density Function of Normal Distribution")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
mean_test = 14
std_dev_test = 2.5
num_cases = 1000
distribution = norm(mean_test, std_dev_test)
p_2_5_to_14 = distribution.cdf(14) - distribution.cdf(2.5)  # P(2.5 < X < 14)
p_below_12 = distribution.cdf(12)  # P(X < 12)
p_above_19 = 1 - distribution.cdf(19)  # P(X > 19)
num_2_5_to_14 = p_2_5_to_14 * num_cases
num_below_12 = p_below_12 * num_cases
num_above_19 = p_above_19 * num_cases
x_vals = np.linspace(mean_test - 4*std_dev_test, mean_test + 4*std_dev_test, 1000)
pdf_vals = distribution.pdf(x_vals)

plt.figure(figsize=(10, 5))
plt.plot(x_vals, pdf_vals, 'r', label="PDF")

x_shade_1 = np.linspace(2.5, 14, 100)
plt.fill_between(x_shade_1, distribution.pdf(x_shade_1), alpha=0.3, color='blue', label="P(2.5 < X < 14)")


x_shade_2 = np.linspace(mean_test - 4*std_dev_test, 12, 100)
plt.fill_between(x_shade_2, distribution.pdf(x_shade_2), alpha=0.3, color='green', label="P(X < 12)")

x_shade_3 = np.linspace(19, mean_test + 4*std_dev_test, 100)
plt.fill_between(x_shade_3, distribution.pdf(x_shade_3), alpha=0.3, color='purple', label="P(X > 19)")

plt.xlabel("Test Score")
plt.ylabel("Density")
plt.title("Normal Distribution with Probability Regions")
plt.legend()
plt.show()

print(f"Number of students who scored between 2.5 and 14: {num_2_5_to_14:.2f}")
print(f"Number of students who scored below 12: {num_below_12:.2f}")
print(f"Number of students who scored above 19: {num_above_19:.2f}")
```
