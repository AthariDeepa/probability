```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from scipy.stats import pearsonr

# Dataset
data = {
    'Study Hours': [2, 4, 6, 8, 1, 5, 7, 3, 9, 10],
    'Exam Scores': [50, 60, 70, 85, 40, 65, 78, 55, 90, 95],
    'Sleep Hours': [8, 7, 6, 5, 9, 6.5, 5, 7.5, 4, 3]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Step 1: Create Scatter Plots
# Scatter Plot: Study Hours vs Exam Scores
plt.figure(figsize=(10, 6))
plt.subplot(1, 3, 1)
plt.scatter(df['Study Hours'], df['Exam Scores'], color='blue')
plt.title('Study Hours vs Exam Scores')
plt.xlabel('Study Hours')
plt.ylabel('Exam Scores')

# Scatter Plot: Sleep Hours vs Exam Scores
plt.subplot(1, 3, 2)
plt.scatter(df['Sleep Hours'], df['Exam Scores'], color='green')
plt.title('Sleep Hours vs Exam Scores')
plt.xlabel('Sleep Hours')
plt.ylabel('Exam Scores')

# Scatter Plot: Study Hours vs Sleep Hours
plt.subplot(1, 3, 3)
plt.scatter(df['Study Hours'], df['Sleep Hours'], color='red')
plt.title('Study Hours vs Sleep Hours')
plt.xlabel('Study Hours')
plt.ylabel('Sleep Hours')

plt.tight_layout()
plt.show()
```
